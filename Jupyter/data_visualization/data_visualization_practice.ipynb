```python
import pandas as pd
df = pd.read_csv("StudentsPerformance.csv")
df[df["reading score"]== 100].plot.bar(x = "test preparation course", y = "reading score")
```

```python
import seaborn as sns
df = pd.read_csv("StudentsPerformance.csv")
sns.lmplot(x = "math score" , y = "reading score" , data = df)
```

### contents and conclusion
Bar chart shows students who scored 100 in reading, separated by whether they completed the test preparation course or did not.
You can compare how many students who scored 100 had completed the prep course versus not completed it.

For the lmplot It shows the relationship between math scores and reading scores.If the regression line slopes upward, it means students who are good at math also tend to be good at reading.If the line were flat or messy, it would suggest no strong relationship between math and reading scores.
